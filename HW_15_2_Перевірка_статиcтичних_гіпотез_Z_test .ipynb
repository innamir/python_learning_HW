1. Запишіть параметри задачі у змінні Python:
    - `sample_mean` - $\bar{x}$, середнє значення в вибірці
    - `population_mean` - $\mu_0$, середнє значення в популяції (тобто те, з яким ми порівнюємо середнє вибіркове значення)
    - `population_std` - $\sigma$, вибіркове стандартне відхилення, яке ми вважаємо, що є рівним ст. відх. популяції, адже маємо достатньо велику вибірку.
    - `sample_size` - $n$, розмір вибірки
    - `alpha` - $\alpha$ рівень значущості
    
    **Увага!** Для обчислення стандартного відхилення маємо скористатись функцією `np.std(your_dataframe, ddof=1)`. Чому так - розбираємо в лекції про t-test.

3. Обчисліть стандартну помилку SE.

7. Прийміть рішення, чи відхиляєте ви гіпотезу $H_0$. Для прийняття рішення зробіть і порівняння z-статистики з критичним значенням, і проаналізуйте p-value.

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
```

```python
# 1. Параметри задачі у змінні Python:
sample_mean = np.mean(data)
population_mean = 20
population_std = np.std(data, ddof=1)
sample_size = len(data)
alpha = 0.05
```

2. Визначаємо гіпотези та тип тесту:
 - Гіпотеза $H_0$:  $\mu_0$ = 20 - продажі залишаються в середньому **$20** на чек

 - Альтернативна гіпотеза $H_a$:  $\mu_A$ > 20 -  середній чек продажів статистично значущо **підвищено**
 - оскільки перевіряємо більше, то тест **правосторонній**.

```python
# 3. Обчислюємо стандартну помилку SE
# SE = σ / √n
standard_error = population_std / np.sqrt(sample_size)
 
print(f"Стандартна помилка, SE = {standard_error}")
```

```python
# 4. Розраховуємо z-статистику.
 # z = (x̄ - μ₀) / SE
z_statistic = (sample_mean - population_mean) / standard_error
print(f"Розрахункова z-статистика, z = {z_statistic}, середній чек вибірки навіть менший за середнє заняення популяції.")
```

```python
# 5. Критичне z-значення з допомогою бібліотеки stats.
z_critical = stats.norm.ppf(1 - alpha)
print(f"Критичне z-значення {z_critical}")
print(f" z < {z_critical:.2f}, гіпотезу H₀ не відхиляємо ")
```

```python
# 6. Обчислюємо p-value.
p_value = 1 - stats.norm.cdf(z_statistic)
print(f"p-value = {p_value}")
```

 7. Приймаємо рішення стосовно гіпотезу $H_0$. 

```python
print(f"p-value = {p_value:.6f}")

print(f"Якщо H₀ правильна -  продажі залишаються в середньому {population_mean} на чек")
print(f"випадково отримати середнє значення чеку ≥ {sample_mean:.2f} становить {p_value:.2f}")
if p_value < 0.001:
    print("Це НАДЗВИЧАЙНО малоймовірно! (менше 0.1%)")
elif p_value < 0.01:
    print("Це дуже малоймовірно! (менше 1%)")
elif p_value < 0.05:
    print("Це малоймовірно! (менше 5%)")
else:
    print("Це цілком можливо випадково, гіпотезу H₀ не відхиляємо")
```
