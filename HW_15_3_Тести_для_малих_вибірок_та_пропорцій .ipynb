# Тести для малих вибірок та пропорцій



**Завдання 1**. E-commerce компанія після редизайну сайту підозрює, що **середній час до покупки (time-to-purchase)** користувачів **збільшився**.
Історично середній час від першого заходу на сайт до покупки становив $\mu_0 = 123.7$ хвилин.

Команда аналітиків випадково обрала дані **7 покупців після редизайну**:
`128, 135, 121, 142, 126, 151, 123`

З допомогою наявних даних зʼясуйте, чи збільшився середній час покупки після редизайну.

Для цього виконайте наступні 6 кроків. Правильне виконання кожного з кроків оцінюється в 1 бал.

3. Оберіть тип тесту та виконайте його (знайдіть критичне значення тесту, статистику тесту та р-значення) будь-яким пасуючим способом, щоб перевірити гіпотезу на рівні значущості $\alpha = 0.10$.
4. Прийміть рішення, чи відхиляєте ви гіпотезу $H_0$ використовуючи p-value.
5. Напишіть висновок: чи справді редизайн сайту подовжив час до покупки?

```python
import numpy as np
import sys
from scipy import stats
import statsmodels.api as sm
from statsmodels.stats.proportion import proportions_ztest, proportion_confint
```

1. Визначаємо гіпотези та тип тесту:
 - Гіпотеза $H_0$:  $\mu_0$ = 123.7 хвилин - середній час від першого заходу на сайт до покупки

 - Альтернативна гіпотеза $H_a$:  $\mu_A$ > 123.7 хвилин -  середній час покупки після редизайну збільшився.
 - оскільки перевіряємо більше, то тест **правосторонній**.

```python
# параметри задачі
data = [128, 135, 121, 142, 126, 151, 123]
sample_mean = np.mean(data)
population_mean = 123.7
population_std = np.std(data, ddof=1)
sample_size = len(data)

alpha = 0.10
# правосторонній t-test
t_stat = (sample_mean - population_mean)/(population_std/ np.sqrt(sample_size))
# критичне значення для правостороннього тесту
df = sample_size - 1
t_crit = stats.t.ppf(1 - alpha, df)
#  p-value
p_value = 1 - stats.t.cdf(t_stat, df)

print(f'Результати "ручного" T-тесту:')
print(f"  t-статистика: {t_stat:.3f}")
print(f"  t-критичне: {t_crit}")
print(f"  P-value: {p_value:.6f}")

# прийняття рішення стосовно гіпотези при α = 0.10
print("ІНТЕРПРЕТАЦІЯ результатів при α = 0.1:")
if p_value < alpha:
    print(f"✅ ВІДХИЛЯЄМО H₀")
    print(f"   p-value ({p_value:.6f}) < α ({alpha})")
    print("   Висновок: Середній час покупки після редизайну збільшився")
else:
    print(f"❌ НЕ ВІДХИЛЯЄМО H₀")
    print(f"   p-value ({p_value:.6f}) ≥ α ({alpha})")
    print("   Висновок: Немає достатньої ймовірності збільшення середнього часу від першого заходу на сайт до покупки")

# прийняття рішення стосовно гіпотези при α = 0.05
alpha = 0.05
print("ІНТЕРПРЕТАЦІЯ результатів при α = 0.05:")
if p_value < alpha:
    print(f"✅ ВІДХИЛЯЄМО H₀")
    print(f"   p-value ({p_value:.6f}) < α ({alpha})")
    print("   Висновок: Середній час покупки після редизайну збільшився")
else:
    print(f"❌ НЕ ВІДХИЛЯЄМО H₀")
    print(f"   p-value ({p_value:.6f}) ≥ α ({alpha})")
    print("   Висновок: Немає достатньої ймовірності збільшення середнього часу від першого заходу на сайт до покупки")
```

```python
# scipy тест

# правосторонній тест при α = 0.10
alpha = 0.10
result_ttest = stats.ttest_1samp(
    data, 
    popmean=population_mean,
    alternative="greater")
print(f"Результати T-тесту")
print(f"  T-statistic: {result_ttest.statistic:.3f}")
print(f"  P-value: {result_ttest.pvalue:.6f}")

# прийняття рішення стосовно гіпотези  α = 0.10
print("ІНТЕРПРЕТАЦІЯ результатів при α = 0.1:")
if result_ttest.pvalue < alpha:
    print(f"✅ ВІДХИЛЯЄМО H₀")
    print(f"   p-value ({result_ttest.pvalue:.6f}) < α ({alpha})")
    print("   Висновок: Середній час покупки після редизайну збільшився")
else:
    print(f"❌ НЕ ВІДХИЛЯЄМО H₀")
    print(f"   p-value ({result_ttest.pvalue:.6f}) ≥ α ({alpha})")
    print("   Висновок: Немає достатньої ймовірності збільшення середнього час від першого заходу на сайт до покупки")

# перевірка гіпотези, якщо рівень значущості буде 5%
alpha = 0.05
# прийняття рішення стосовно гіпотези
print("ІНТЕРПРЕТАЦІЯ результатів при α = 0.1:")
if result_ttest.pvalue < alpha:
    print(f"✅ ВІДХИЛЯЄМО H₀")
    print(f"   p-value ({result_ttest.pvalue:.6f}) < α ({alpha})")
    print("   Висновок: Середній час покупки після редизайну збільшився")
else:
    print(f"❌ НЕ ВІДХИЛЯЄМО H₀")
    print(f"   p-value ({result_ttest.pvalue:.6f}) ≥ α ({alpha})")
    print("   Висновок: Немає достатньої ймовірності збільшення середнього час від першого заходу на сайт до покупки")
```

**Завдання 2.**

До спеціальної рекламної кампанії **23%** дорослих упізнавали логотип компанії. Після завершення кампанії відділ маркетингу провів опитування: з **1200** випадково відібраних дорослих **311** упізнали логотип.

Перевірте на рівні значущості **3%** ($\alpha=0.03$), чи дають ці дані достатні підстави стверджувати, що **тепер більше ніж 23%** дорослих упізнають логотип компанії. Для розвʼязку використовуйте бібліотеку `statsmodels`.

Зробіть висновок, чи зросла впізнаваність логотипу.

Додатково, обчисліть довірчий інтревал на заданому рівні значущості і проінтерпретуйте текстом - як він додатково пояснює прийняте нами рішення?

Розв'язок: 

Визначаємо гіпотези та тип тесту:
 - Гіпотеза $H_0$:  $\mu_0$ = 23% дорослих упізнають логотип компанії після завершення кампанії відділ маркетингу

 - Альтернативна гіпотеза $H_a$:  $\mu_A$ > 23% дорослих упізнають логотип компанії після завершення кампанії відділ маркетингу
 - оскільки перевіряємо більше, то тест **правосторонній**.

```python
import statsmodels.api as sm
from statsmodels.stats.proportion import proportions_ztest, proportion_confint

count = 311 
nobs = 1200    
p0 = 0.23 
alpha = 0.03

# z-тест для пропорції
stat, pval = proportions_ztest(count, nobs, p0, alternative='larger')
confint = proportion_confint(count, nobs, alpha=0.03, method='normal')

print("Тест для однієї пропорції")
print(f"Вибіркова пропорція = {count/nobs:.4f}")
print(f"Z-статистика = {stat:.3f}")
print(f"p-value = {pval:.4f}")
print(f"97% довірчий інтервал: {confint[0]:.4f}, {confint[1]:.4f}")

if pval < alpha:
    print(f"Оскільки p-value = {pval:.3f} < {alpha}, відхиляємо H₀.")
    print(f"Є статистично значущі докази, що після рекламної кампанії більше ніж  23% дорослих упізнають логотип компанії.")
else:
    print(f"Оскільки p-value = {pval:.3f} > {alpha},   не відхиляємо H₀.")
    print(f"Недостаьньо доказів стверджувати, що після рекламної кампанії більше ніж  23% дорослих упізнають логотип компанії.")

print(f"\nІнтерпретація довірчого інтервалу:")
print(f"97% довірчий інтервал для істинної частки становить від {confint[0]*100:.2f}% до {confint[1]*100:.2f}%.")

if confint[0] > p0:
    print("Оскільки весь інтервал лежить вище 23%, це додатково підтверджує висновок про зростання впізнаваності логотипу.")
else:
    print("Оскільки 23% потрапляє в інтервал або близько до його межі, інтервал не дає такого сильного підтвердження.")
```
